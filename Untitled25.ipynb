```python
#kerbside recycling(hovered over) and year(x axis) and residential/non residential (z axis), kerbside_recycling_total_collected_tonnes','kerbside_recycling_total_recycled_tonnes'
import pandas as pd
import plotly.express as px

# Create a bar plot using Plotly Express
fig = px.bar(df, x='financial_year', y='kerbside_recycling_total_collected_tonnes',
             color='kerbside_recycling_service_frequency',
             labels={'kerbside_recycling_total_collected_tonnes': 'Total Collected Tonnes'},
             title='Kerbside Recycling Over Financial Years')

fig.update_layout(xaxis_title='Financial Year')
fig.update_yaxes(title_text='Total Collected Tonnes')

fig.show()
```

```python
# List of recycling types you want to visualize
recycling_types = ['kerbside_recycling_aluminimum_tonnes', 'kerbside_recycling_paper_all_collected_tonnes',
                   'kerbside_recycling_glass_all_collected_tonnes', 'kerbside_recycling_plastics_all_collected_tonnes']

# Create separate bar plots for each recycling type using Plotly Express
for recycling_type in recycling_types:
    fig = px.bar(df, x='financial_year', y=recycling_type,
                 title=f'{recycling_type.replace("_", " ").title()} Over the Years',
                 labels={recycling_type: 'Collected Tonnes'})
    fig.update_xaxes(rotation=45)
    fig.show()
```
